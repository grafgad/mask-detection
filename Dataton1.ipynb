```python
# загружаем библиотеки
from roboflow import Roboflow
from IPython.display import Image

# читаем apikey для работы с Roboflow
with open('apikey.txt') as f:
    apikey = f.read()
    
# загружаем модель
rf = Roboflow(api_key=apikey)
project = rf.workspace().project("masks-detection-kwy1r")
model = project.version(3).model
```

```python
# infer on a local image
#print(model.predict("mask.jpg", confidence=40, overlap=30).json())
# infer on an image hosted elsewhere
# print(model.predict("URL_OF_YOUR_IMAGE", hosted=True, confidence=40, overlap=30).json())

# visualize your prediction
model.predict("mask.jpg", confidence=40, overlap=30).save("prediction_mask.jpg") # проверяем наличие маски
Image('prediction_mask.jpg') # выводим изображение
```
