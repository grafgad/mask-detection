```python
# загружаем библиотеки
from roboflow import Roboflow
from IPython.display import Image

# читаем apikey для работы с Roboflow
with open('apikey.txt') as f:
    apikey = f.read()
    
# загружаем модель
rf = Roboflow(api_key=apikey)
project = rf.workspace().project("masks-detection-kwy1r")
model = project.version(3).model
```

```python
# напишем функцию, которая принимает фотографию и возвращает результат проверки
def check_mask_on_photo(photo: str, prediction: str):
    
    model.predict(photo, confidence=40, overlap=30).save(prediction) # проверяем наличие маски
    return Image(prediction) # выводим изображение
```

```python
# Проверка на фото с маской
check_mask_on_photo("mask.jpg", "prediction_mask.jpg")
```

```python
# Проверка на фото с маской
check_mask_on_photo("mask2.jpg", "prediction_mask2.jpg")
```

```python
# Проверка на фото с маской
check_mask_on_photo("mask4.jpg", "prediction_mask4.jpg")
```

```python
# Проверка на фото без маски
check_mask_on_photo("no_mask2.jpg", "prediction_no_mask.jpg")
```
